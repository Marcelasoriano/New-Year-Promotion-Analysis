# 3. Ingresos Estimados para Cada Escenario

```python
# Escenarios de apuestas
escenario_conservador = (0.80, 20, 0.20, 40)  # 80% apuestan 20Q, 20% apuestan 40Q
escenario_moderado = (0.60, 20, 0.40, 60)     # 60% apuestan 20Q, 40% apuestan 60Q
escenario_optimista = (0.40, 20, 0.60, 80)    # 40% apuestan 20Q, 60% apuestan 80Q

# Tasa de participación estimada
tasa_participacion = 0.05  # 5%

# Estimación de la población de la Ciudad de Guatemala en el rango de edad de 18 a 65+ años
poblacion_total_guatemala = 18000000  # Estimación total de la población de Guatemala
porcentaje_ciudad_guatemala = 0.10  # Porcentaje de la población en la Ciudad de Guatemala
porcentaje_rango_edad = 0.50  # Porcentaje en el rango de edad de 18 a 65+ años
poblacion_estimada_ciudad_guatemala = poblacion_total_guatemala * porcentaje_ciudad_guatemala * porcentaje_rango_edad

# Número estimado de participantes en la campaña
participantes_estimados = poblacion_estimada_ciudad_guatemala * tasa_participacion

# Función para calcular ingresos estimados para cada escenario con la cantidad estimada de participantes
def calcular_ingresos_con_participantes(porcentaje_minimo, apuesta_minima, porcentaje_mayor, apuesta_promedio_mayor, participantes):
    ingresos_minimos = participantes * porcentaje_minimo * apuesta_minima
    ingresos_mayores = participantes * porcentaje_mayor * apuesta_promedio_mayor
    return ingresos_minimos + ingresos_mayores

# Calcular ingresos estimados para cada escenario con la cantidad estimada de participantes
ingresos_conservador_con_participantes = calcular_ingresos_con_participantes(*escenario_conservador, participantes_estimados)
ingresos_moderado_con_participantes = calcular_ingresos_con_participantes(*escenario_moderado, participantes_estimados)
ingresos_optimista_con_participantes = calcular_ingresos_con_participantes(*escenario_optimista, participantes_estimados)

# Imprimir resultados con etiquetas
resultados_con_labels = {
    "Ingresos Conservador (Q)": ingresos_conservador_con_participantes,
    "Ingresos Moderado (Q)": ingresos_moderado_con_participantes,
    "Ingresos Optimista (Q)": ingresos_optimista_con_participantes
}

resultados_con_labels
```

# 4. Cálculo del ROI

```python
# Ingresos estimados para cada escenario (en Quetzales)
ingresos_escenario_conservador = ingresos_conservador_con_participantes
ingresos_escenario_moderado = ingresos_moderado_con_participantes
ingresos_escenario_optimista = ingresos_optimista_con_participantes

# Conversión de costos a Quetzales (asumiendo una tasa de cambio: 1 USD = 7.5 Q)
tasa_cambio_usd_a_q = 7.5
costo_total_campana = 3490  # Costo total de la campaña en USD
costo_total_campana_q = costo_total_campana * tasa_cambio_usd_a_q

# Función para calcular el ROI
def calcular_roi(ingresos, costo_campana):
    ganancia = ingresos - costo_campana
    roi = (ganancia / costo_campana) * 100
    return roi

# ROI para cada escenario
roi_conservador = calcular_roi(ingresos_escenario_conservador, costo_total_campana_q)
roi_moderado = calcular_roi(ingresos_escenario_moderado, costo_total_campana_q)
roi_optimista = calcular_roi(ingresos_escenario_optimista, costo_total_campana_q)

# Imprimir resultados con etiquetas
resultados_roi = {
    "ROI Conservador (%)": roi_conservador,
    "ROI Moderado (%)": roi_moderado,
    "ROI Optimista (%)": roi_optimista
}

resultados_roi
```

# 7. Análisis de Sensibilidad para la Distribución de Apuestas

## Visualización

```python
# Funciones necesarias
def calcular_ingresos_con_participantes(porcentaje_minimo, apuesta_minima, porcentaje_mayor, apuesta_promedio_mayor, participantes):
    ingresos_minimos = participantes * porcentaje_minimo * apuesta_minima
    ingresos_mayores = participantes * porcentaje_mayor * apuesta_promedio_mayor
    return ingresos_minimos + ingresos_mayores

def calcular_roi(ingresos, costo_campana):
    ganancia = ingresos - costo_campana
    roi = (ganancia / costo_campana) * 100
    return roi

def calcular_ingresos_y_roi_por_tasa_participacion(tasa_participacion, escenario, poblacion_ciudad, costo_campana_q):
    participantes = poblacion_ciudad * tasa_participacion
    ingresos = calcular_ingresos_con_participantes(*escenario, participantes)
    roi = calcular_roi(ingresos, costo_campana_q)
    return ingresos, roi

# Parámetros iniciales
tasa_participacion_original = 0.05  # 5%
tasa_cambio_usd_a_q = 7.5
costo_total_campana = 3490  # Costo total de la campaña en USD
costo_total_campana_q = costo_total_campana * tasa_cambio_usd_a_q
poblacion_estimada_ciudad_guatemala = 18000000 * 0.10 * 0.50  # Calculado previamente

# Escenario moderado
escenario_moderado = (0.60, 20, 0.40, 60)  # 60% apuesta 20Q, 40% apuesta 60Q
ingresos_escenario_moderado = calcular_ingresos_con_participantes(*escenario_moderado, poblacion_estimada_ciudad_guatemala * tasa_participacion_original)
roi_escenario_moderado = calcular_roi(ingresos_escenario_moderado, costo_total_campana_q)

# Nuevo escenario de sensibilidad
nuevo_escenario_sensibilidad = (0.50, 20, 0.50, 60)  # 50% apuestan 20Q, 50% apuestan 60Q
ingresos_nuevo_escenario, roi_nuevo_escenario = calcular_ingresos_y_roi_por_tasa_participacion(tasa_participacion_original, nuevo_escenario_sensibilidad, poblacion_estimada_ciudad_guatemala, costo_total_campana_q)

# Preparando datos para visualización
datos_nuevo_escenario = {
    "Escenario": ["Moderado", "Nuevo Escenario"],
    "Ingresos (Q)": [ingresos_escenario_moderado, ingresos_nuevo_escenario],
    "ROI (%)": [roi_escenario_moderado, roi_nuevo_escenario]
}

df_nuevo_escenario = pd.DataFrame(datos_nuevo_escenario)

# Graficando los datos
plt.figure(figsize=(12, 6))

# Ingresos
plt.subplot(1, 2, 1)
sns.barplot(x="Escenario", y="Ingresos (Q)", data=df_nuevo_escenario, palette="magma")
plt.title("Comparación de Ingresos")
plt.xlabel("Escenario")
plt.ylabel("Ingresos (Q)")

# ROI
plt.subplot(1, 2, 2)
sns.barplot(x="Escenario", y="ROI (%)", data=df_nuevo_escenario, palette="cividis")
plt.title("Comparación de ROI")
plt.xlabel("Escenario")
plt.ylabel("ROI (%)")

plt.tight_layout()
plt.show()
```

### El ROI está expresado en porcentaje (%). Este valor representa la rentabilidad de la campaña en términos de retorno sobre la inversión. El cálculo del ROI toma en cuenta tanto los ingresos (en Quetzales) como los costos de la campaña (originalmente en dólares estadounidenses, pero convertidos a Quetzales para el cálculo).

### Moderado:
- #### El ROI moderado refleja la rentabilidad estimada en un escenario intermedio en términos de apuestas. Indica cuánto se espera que rinda la inversión en un escenario equilibrado en el que algunos participantes hacen apuestas más bajas y otros hacen apuestas más altas.

### Optimista:
- #### El ROI optimista muestra la rentabilidad anticipada en el escenario más optimista, donde la mayoría de los participantes hacen apuestas más altas. Un ROI optimista alto indicaría que la inversión inicial tiene el potencial de generar un retorno significativo en ganancias.
